```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.boxplot(df[['appid','achievements']], labels=['appid','penghargaan'])
plt.title('GRAFIK BOXPLOT')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```

```python
sns.heatmap(data=df.corr())
plt.title('GRAFIK HEATMAP STEAM')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```

```python
df2 = pd.read_csv('players_fifa.csv')
df2.head()
```

```python
df2.info()
```

```python
plt.figure(figsize=(10,5))
plt.title('GRAFIK PIE FIFA')
plt.pie(df2['position'].value_counts(), labels=['gk','cf','mf','df'], startangle=90, autopct='%.1f%%', explode=(0.1,0.1,0.1,0.1))
plt.legend(loc='lower right')
plt.show()
```

```python
df2['position'].value_counts().plot.bar()
plt.title('GRAFIK BATANG FIFA')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('GRAFIK BATANG FIFA')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.hist(df2['position'], label ='Posisi')
plt.hist(df2['age'], label ='Umur')
plt.legend()
plt.xticks(rotation=60)
plt.show()
```
